```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sl
```

```python
df = pd.read_csv('bench_20231116_141413_timestep_128_psrn_by_tol.txt')
df.head()
```

```python
df["comp_ratio"] = df["oriSize"]/df["compSize"]
```

```python
df
```

```python
tol = df['tol']
comp_ratio = df['comp_ratio']
psnr = df['psnr']

# Create the first graph (tol vs comp_ratio)
plt.figure(figsize=(8, 6))
plt.plot(tol, comp_ratio, marker='o', linestyle='-', color='b')
plt.title('tol vs comp_ratio')
plt.xlabel('tol')
plt.ylabel('comp_ratio')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.grid(True)
plt.savefig('./tol_VS_comp_ratio.png')

# Create the second graph (tol vs psnr)
plt.figure(figsize=(8, 6))
plt.plot(tol, psnr, marker='o', linestyle='-', color='r')
plt.title('tol vs psnr')
plt.xlabel('tol')
plt.ylabel('psnr')
plt.xscale('log')  # Set x-axis to logarithmic scale
plt.grid(True)
plt.savefig('./tol_VS_psnr_ratio.png')
```

```python
# Create a single graph with two y-axes
fig, ax1 = plt.subplots(figsize=(8, 6))

# First plot: tol vs comp_ratio on the left y-axis
ax1.plot(df['tol'], df['comp_ratio'], marker='s', linestyle='-', color='b', label='compression ratio')
ax1.set_xlabel('tol',fontsize=14)
ax1.set_ylabel('compression ratio',fontsize=14)
ax1.set_xscale('log')  # Set x-axis to logarithmic scale
ax1.grid(True)
ax1.tick_params(axis='y')
ax1.legend(loc='upper left')

# Create a second y-axis on the right side
ax2 = ax1.twinx()

# Second plot: tol vs psnr on the right y-axis
ax2.plot(df['tol'], df['psnr'], marker='*', linestyle='-', color='r', label='psnr')
ax2.set_ylabel('psnr',fontsize=14)
ax2.tick_params(axis='y')
ax2.legend(loc='upper right')

# Set title
# plt.title('tol vs comp_ratio and tol vs psnr')

# Save the combined graph
plt.savefig('./combined_graph.png')

# Show the graph
plt.show()
```
